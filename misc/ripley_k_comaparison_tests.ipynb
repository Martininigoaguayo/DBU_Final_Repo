```python
from libs.prefect_helpers import *
from libs.data_manipulation import *
from libs.feature_generation import *
from libs.dim_reduction import *
from libs.football_plots import *
from libs.clustering import *
from libs.convex_hull import *
```

```python
data = compile_team_tracking_data("data/H_EURO2020", "DEN")
one_match = extract_one_match(data, 1)


ripley_k_multiclass_whole_match = ripley_k_multiclass_by_indices(one_match, one_match.iloc[::48].index)
ripley_k_whole_match = ripley_k_by_indices(one_match, one_match.iloc[::48].index)
```

```python
multiclass_ripk_PCA = PCAObject(ripley_k_multiclass_whole_match)
ripk_PCA = PCAObject(ripley_k_whole_match)
```

```python
multiclass_ripk = multiclass_ripk_PCA.transform(ripley_k_multiclass_whole_match)
ripk = ripk_PCA.transform(ripley_k_whole_match)
```

```python
plt.scatter(multiclass_ripk[:,0], multiclass_ripk[:,1])
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA of Multi-class ripleys k")
plt.savefig("Ripleysk_k_plots/ripk_multiclass_PCA")
```
